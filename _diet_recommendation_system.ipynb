```python
# Encoding categorical features
le_workout = LabelEncoder()
le_disease = LabelEncoder()
le_allergy = LabelEncoder()
le_diet = LabelEncoder()
le_food = LabelEncoder()
df['Workout Type'] = le_workout.fit_transform(df['Workout Type'])
df['Disease'] = le_disease.fit_transform(df['Disease'])
df['Allergies'] = le_allergy.fit_transform(df['Allergies'])
df['Recommended Diet'] = le_diet.fit_transform(df['Recommended Diet'])
df['Recommended Food'] = le_food.fit_transform(df['Recommended Food'])
```

```python
# Selecting features and target
X = df[['Age', 'Weight', 'Height', 'Workout Type', 'Disease', 'Allergies']]
y = df[['Recommended Diet', 'Recommended Food']]
```

```python
y = df[['Recommended Diet', 'Recommended Food']]
```

```python
y_diet = df['Recommended Diet']
y_food = df['Recommended Food']
```

```python
X_train, X_test, y_train_diet, y_test_diet, y_train_food, y_test_food = train_test_split(
    X, y_diet, y_food, test_size=0.2, random_state=42
)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(n_neighbors=3),
    'Naive Bayes': GaussianNB()
}
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
for name, model in models.items():
    model.fit(X_train, y_train_diet)  # Train for diet recommendations
    y_pred_diet = model.predict(X_test)

    model.fit(X_train, y_train_food)  # Train for food recommendations
    y_pred_food = model.predict(X_test)

    accuracy_diet = accuracy_score(y_test_diet, y_pred_diet)
    accuracy_food = accuracy_score(y_test_food, y_pred_food)

    print(f'{name} Model Accuracy - Diet: {accuracy_diet * 100:.2f}%')
    print(f'{name} Model Accuracy - Food: {accuracy_food * 100:.2f}%')
```

```python
# Training and evaluating models
for name, model in models.items():
    model.fit(X_train, y_train_diet)
    y_pred_diet = model.predict(X_test)

    model.fit(X_train, y_train_food)
    y_pred_food = model.predict(X_test)

    accuracy_diet = accuracy_score(y_test_diet, y_pred_diet)
    accuracy_food = accuracy_score(y_test_food, y_pred_food)

    print(f'{name} Model Accuracy - Diet: {accuracy_diet * 100:.2f}%')
    print(f'{name} Model Accuracy - Food: {accuracy_food * 100:.2f}%')

    print("Diet Recommendation Classification Report:")
    print(classification_report(y_test_diet, y_pred_diet))

    print("\nFood Recommendation Classification Report:")
    print(classification_report(y_test_food, y_pred_food))
```

```python
# Function to recommend diet food based on user input
def recommend_diet_food():
    print("\n=== Diet Recommendation System ===")
    age = int(input("Enter your age: "))
    weight = float(input("Enter your weight (kg): "))
    height = float(input("Enter your height (cm): "))
    workout_type = input("Enter your workout type (Strength/Cardio/Yoga): ").strip().capitalize()
    disease = input("Enter your disease condition (Diabetes/Hypertension/Cholesterol/None): ").strip().capitalize()
    allergies = input("Do you have any food allergies? (Yes/No): ").strip().capitalize()

    workout_encoded = le_workout.transform([workout_type])[0]
    disease_encoded = le_disease.transform([disease])[0]
    allergies_encoded = le_allergy.transform([allergies])[0]
    input_data = np.array([[age, weight, height, workout_encoded, disease_encoded, allergies_encoded]])

    best_model = max(models.items(), key=lambda x: accuracy_score(y_test_diet, x[1].predict(X_test)))[1]
    predicted_diet = best_model.predict(input_data)
    best_model_food = max(models.items(), key=lambda x: accuracy_score(y_test_food, x[1].predict(X_test)))[1]
    predicted_food = best_model_food.predict(input_data)

    diet_label = le_diet.inverse_transform([predicted_diet[0]])[0]
    food_label = le_food.inverse_transform([predicted_food[0]])[0]

    print(f'\nRecommended Diet: {diet_label}')
    print(f'Recommended Food: {food_label}')
```

```python
# Run the recommendation function
recommend_diet_food()
```
